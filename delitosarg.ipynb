```python
import pandas as pd
import matplotlib.pyplot as plt

# Extraer las dos columnas para la gráfica
columna1 = df_delitos['robos_hechos_buenos_aires']
columna2 = df_delitos['indice_tiempo']

# Crear la gráfica
plt.figure(figsize=(10, 10))  # Ajusta el tamaño de la figura según sea necesario
plt.plot(columna1, columna2, marker='o', linestyle='-')  # Graficar los datos

# Personalizar la gráfica (opcional)
plt.title('Delitos en Buenos Aires')
plt.xlabel('Columna 1')
plt.ylabel('Columna 2')
plt.grid(True)  # Mostrar cuadrícula en la gráfica (opcional)

# Mostrar la gráfica
plt.show()
```
